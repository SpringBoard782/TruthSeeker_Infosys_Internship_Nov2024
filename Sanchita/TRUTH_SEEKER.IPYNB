```python
# Display count of each unique value in the 'label' column
label_counts = combined_df['label'].value_counts()
print(label_counts)
```

```python
# Check for null values in each column
print(combined_df.isnull().sum())
```

```python
# Drop rows where either 'text' or 'title' column have NaN values
combined_df = combined_df.dropna(subset=['text', 'title'])
```

```python
print(combined_df.isnull().sum())
```

```python
combined_df.head()
```

```python
# Count the total number of duplicate rows
duplicate_count = combined_df.duplicated().sum()
print(f"Total number of duplicate rows: {duplicate_count}")

# Display the duplicate rows
duplicate_rows = combined_df[combined_df.duplicated()]
print("Duplicate rows:")
print(duplicate_rows)
```

```python
combined_df.drop_duplicates(keep='first', inplace=True)
combined_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count the occurrences of each label (fake and real)
label_counts = combined_df['label'].value_counts().reset_index()
label_counts.columns = ['label', 'count']

# Define a custom palette using light green and blue
custom_palette = ['#90EE90', '#4682B4']  # Light Green and Blue (Hex Codes)

# Create a bar plot using the custom palette
plt.figure(figsize=(8, 5))
sns.barplot(x='label', y='count', data=label_counts, palette=custom_palette, hue='label', dodge=False)

# Set labels and title
plt.xlabel('Label (0 = Real, 1 = Fake)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Distribution of Fake and Real News', fontsize=14, fontweight='bold')

# Add gridlines and style
sns.despine()
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
combined_df['title_length']=combined_df['title'].astype(str).apply(len)
combined_df['text_length']=combined_df['text'].astype(str).apply(len)
```

# EDA

```python
# Create a column for word count
combined_df['text_word_count'] = combined_df['text'].apply(lambda x: len(str(x).split()))

# Plot distribution of word counts
plt.figure(figsize=(8, 5))
sns.histplot(combined_df['text_word_count'], bins=50, kde=True, color='blue')
plt.xlabel("Word Count in Text")
plt.ylabel("Frequency")
plt.title("Distribution of Word Count in Text", fontweight='bold')  # Make title bold
plt.show()
```

```python
# Create a column for title length
combined_df['title_length'] = combined_df['title'].apply(lambda x: len(str(x)))

# Plot distribution of title length
plt.figure(figsize=(8, 5))
sns.histplot(combined_df['title_length'], bins=30, kde=True, color='#5A9')  # Teal color
plt.xlabel("Title Length", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Distribution of Title Length", fontsize=14, fontweight='bold')

# Style adjustments
sns.despine()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plot word count distribution by label
plt.figure(figsize=(10, 6))
sns.histplot(data=combined_df, 
             x='text_word_count', 
             hue='label', 
             multiple='stack', 
             bins=50, 
             palette='Set1')
plt.xlabel("Word Count in Text")
plt.ylabel("Frequency")
plt.title("Word Count Distribution by Label", fontweight='bold')  # Make title bold
plt.show()
```

```python
# Plot title length distribution by label
plt.figure(figsize=(10, 6))
sns.histplot(data=combined_df, 
             x='title_length', 
             hue='label', 
             multiple='stack', 
             bins=30, 
             palette='Set2')
plt.xlabel("Title Length")
plt.ylabel("Frequency")
plt.title("Title Length Distribution by Label", fontweight='bold')  # Make title bold
plt.show()
```

```python
from wordcloud import WordCloud

# Word cloud for fake news
fake_news = combined_df[combined_df['label'] == 1]['text'].dropna()
fake_text = " ".join(fake_news)
wordcloud_fake = WordCloud(width=800, height=400, background_color='black', colormap='Reds').generate(fake_text)

# Display fake news word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_fake, interpolation='bilinear')
plt.axis('off')
plt.title("Most Common Words in Fake News")
plt.show()
```

```python
# Word cloud for real news
real_news = combined_df[combined_df['label'] == 0]['text'].dropna()
real_text = " ".join(real_news)
wordcloud_real = WordCloud(width=800, height=400, background_color='white', colormap='Spectral').generate(real_text)

# Display real news word cloud
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud_real, interpolation='bilinear')
plt.axis('off')
plt.title("Most Common Words in Real News")
plt.show()
```

```python
# Display correlation heatmap (only works if additional numeric features are present)
plt.figure(figsize=(8, 6))
sns.heatmap(combined_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
