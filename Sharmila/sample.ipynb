```python
# Function to map NLTK POS tags to WordNet POS tags
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return 'a'  # adjective
    elif tag.startswith('V'):
        return 'v'  # verb
    elif tag.startswith('N'):
        return 'n'  # noun
    elif tag.startswith('R'):
        return 'r'  # adverb
    else:
        return 'n'  # default to noun
```

### Text Preprocessing Techniques
#### -> Lowercasing
#### -> Removing Punctuation and Special Characters
#### -> Tokenization
#### -> Stopword Removal
#### -> Lemmatization
#### -> POS Tagging

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    words = word_tokenize(text)
    tagged_words = pos_tag(words)
    processed_words = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in tagged_words
        if word not in stop_words and len(word) > 2
    ]
    return ' '.join(processed_words)
```

```python
def process_large_csv(file_path, chunk_size=1000):
    processed_chunks = []
    for chunk in pd.read_csv(file_path, chunksize=chunk_size):
        chunk['title'] = chunk['title'].fillna("missing data")
        chunk['text'] = chunk['text'].fillna("missing data")
        chunk['processed_title'] = chunk['title'].apply(preprocess_text)
        chunk['processed_text'] = chunk['text'].apply(preprocess_text)
        processed_chunks.append(chunk)
    return pd.concat(processed_chunks, ignore_index=True)
```

```python
texts = sampled_data['processed_text'].astype(str)
labels = sampled_data['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)
```

```python
tokenized_sentences = [text.split() for text in x_train]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(x_train)
```

```python
embedding_dim = 100
word2vec_model = Word2Vec(sentences=tokenized_sentences, vector_size=embedding_dim, window=5, min_count=1, workers=4)


word2vec_model.save("/content/drive/My Drive/Colab Notebooks/word2vec.model")
```

```python
word_index = tokenizer.word_index
vocab_size = len(word_index) + 1
```

```python
x_train_seq = tokenizer.texts_to_sequences(x_train)
x_test_seq = tokenizer.texts_to_sequences(x_test)
```

```python
embedding_matrix = np.zeros((vocab_size, embedding_dim))

for word, i in word_index.items():
    if word in word2vec_model.wv:
        embedding_matrix[i] = word2vec_model.wv[word]

zero_rows = np.sum(np.all(embedding_matrix == 0, axis=1))
print(f"Number of zero rows (OOV words or padding): {zero_rows}")
```

```python
embedding_matrix
```

```python
max_length = 500
x_train_padded = pad_sequences(x_train_seq, maxlen=max_length, padding='post', truncating='post')
x_test_padded = pad_sequences(x_test_seq, maxlen=max_length, padding='post', truncating='post')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

model = Sequential([
    Embedding(
        input_dim=vocab_size,
        output_dim=embedding_dim,
        weights=[embedding_matrix],
        input_length=max_length,
        trainable=False
    ),
    LSTM(128, return_sequences=False),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])


model.compile(
    optimizer=Adam(learning_rate=1e-3),
    loss='binary_crossentropy',
    metrics=['accuracy']
)


early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)


history = model.fit(
    x_train_padded,
    y_train,
    validation_split=0.2,
    epochs=20,
    batch_size=64,
    callbacks=[early_stopping]
)
```

```python
loss, accuracy = model.evaluate(x_test_padded, y_test)
print(f"Test Loss: {loss*100}")
print(f"Test Accuracy: {accuracy*100}")
```

```python
texts = df['processed_text'].astype(str)
labels = df['label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(texts, labels, test_size=0.2, random_state=42)
```

```python
max_length = 5500
x_train_pad = pad_sequences(x_train_seq, maxlen=max_length)
x_test_pad = pad_sequences(x_test_seq, maxlen=max_length)
```
