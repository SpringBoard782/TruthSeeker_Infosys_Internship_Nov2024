```python
from collections import Counter
import re

def get_common_words(text_data, num_words=10):
    words = ' '.join(text_data).lower()
    words = re.findall(r'\b\w+\b', words)
    return Counter(words).most_common(num_words)

print("Most common words in fake news titles:")
print(get_common_words(df[df['label'] == 1]['title']))

print("Most common words in real news titles:")
print(get_common_words(df[df['label'] == 0]['title']))
```

```python
# Function to map NLTK POS tags to WordNet POS tags
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return 'a'  # adjective
    elif tag.startswith('V'):
        return 'v'  # verb
    elif tag.startswith('N'):
        return 'n'  # noun
    elif tag.startswith('R'):
        return 'r'  # adverb
    else:
        return 'n'  # default to noun
```

### Below text preprocessing code involves:
#### -> Lowercasing
#### -> Removing Punctuation and Special Characters
#### -> Tokenization
#### -> Stopword Removal
#### -> Lemmatization
#### -> POS Tagging

```python
# Preprocessing function
def preprocess_text(text):
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize
    words = word_tokenize(text)
    # POS tagging
    tagged_words = pos_tag(words)
    # Lemmatize and remove stopwords
    processed_words = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in tagged_words
        if word not in stop_words and len(word) > 2
    ]
    return ' '.join(processed_words)
```

```python
# Efficiently process large CSV file
def process_large_csv(file_path, chunk_size=1000):
    processed_chunks = []
    for chunk in pd.read_csv(file_path, chunksize=chunk_size):
        chunk['title'] = chunk['title'].fillna("missing data")  # Handle missing values
        chunk['text'] = chunk['text'].fillna("missing data")  # Handle missing values
        chunk['processed_title'] = chunk['title'].apply(preprocess_text)
        chunk['processed_text'] = chunk['text'].apply(preprocess_text)
        processed_chunks.append(chunk)
    return pd.concat(processed_chunks, ignore_index=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=20000, min_df=5, max_df=0.9)
```

```python
X = tfidf.fit_transform(df['processed_text'])
y = df['label']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred)*100)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': [0.00001, 0.0001, 0.001, 0.01, 0.1]}

grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
best_model = grid_search.best_estimator_

print("Training accuracy:", best_model.score(X_train, y_train)*100)
print("Validation accuracy:", best_model.score(X_test, y_test)*100)

y_pred = best_model.predict(X_test)
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
print("All parameters of best model:", best_model.get_params())
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform

param_dist = {'alpha': uniform(0.001, 100)}
random_search = RandomizedSearchCV(model, param_distributions=param_dist, n_iter=50, cv=5, random_state=42)
random_search.fit(X_train, y_train)
```

```python
print("Best parameters:", grid_search.best_params_)
print("Best cross-validation accuracy:", grid_search.best_score_*100)
```
