```python
from collections import Counter
import re

def get_common_words(text_data, num_words=10):
    words = ' '.join(text_data).lower()
    words = re.findall(r'\b\w+\b', words)
    return Counter(words).most_common(num_words)

print("Most common words in fake news titles:")
print(get_common_words(df[df['label'] == 1]['title']))

print("Most common words in real news titles:")
print(get_common_words(df[df['label'] == 0]['title']))
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
df
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

### Below text preprocessing code involves:
#### -> Lowercasing
#### -> Removing Punctuation and Special Characters
#### -> Tokenization
#### -> Stopword Removal
#### -> Lemmatization

```python
def preprocess_text(text):
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize
    words = word_tokenize(text)
    # Remove stopwords and lemmatize
    processed_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words and len(word) > 2]
    return ' '.join(processed_words)

#preprocessing to both 'title' and 'text' columns
df['processed_title'] = df['title'].apply(preprocess_text)
df['processed_text'] = df['text'].apply(preprocess_text)
```
