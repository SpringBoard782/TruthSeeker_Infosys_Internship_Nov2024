```python
import pandas as pd

# Load the CSV files into pandas DataFrames
fake_df = pd.read_csv('Fake.csv')
true_df = pd.read_csv('True.csv')

# Concatenate the DataFrames into a single DataFrame
data = pd.concat([fake_df, true_df])

# Print some info to verify the data was loaded correctly
print(data.head())
print(data.shape)
```

```python

# Function to lowercase all text columns
def lowercase_text_columns(df):
    text_cols = df.select_dtypes(include=['object']).columns  # Select object (string) columns
    for col in text_cols:
        df[col] = df[col].str.lower()
    return df

# Lowercase the text in both DataFrames
fake_df = lowercase_text_columns(fake_df)
true_df = lowercase_text_columns(true_df)

# Now you can use the lowercased DataFrames (fake_df and true_df) for further processing
# For example, save them back to new CSV files
fake_df.to_csv('fake_lower.csv', index=False)
true_df.to_csv('true_lower.csv', index=False)
```

```python
combined_news = pd.concat([true_df, fake_df], ignore_index=True)
print(combined_news.head())
```

```python
import re

combined_news['cleaned_text'] = combined_news['text'].apply(lambda x: re.sub(r'[^\w\s]', '', str(x)))

print(combined_news.head())
```

```python
## Removing stop words
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words=set(stopwords.words('english'))
```

```python

def remove_stop_words(text):

    filtered_words = [word for word in text.split() if word.lower() not in stop_words]
    return ' '.join(filtered_words)


combined_news['text'] = combined_news['text'].apply(remove_stop_words)

print(combined_news[['text', 'label']].head())
```

```python
from contractions import contractions_dict
```

```python
def expand_contractions(text):
    # Regular expression to match contractions in text
    contraction_pattern = re.compile(r'\b(' + '|'.join(contractions_dict.keys()) + r')\b')

    # Substitute each contraction in text with its expanded form
    expanded_text = contraction_pattern.sub(lambda x: contractions_dict[x.group()], text)
    return expanded_text
```
