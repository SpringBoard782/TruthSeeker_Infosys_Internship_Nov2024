```python
import re
import pandas as pd
import spacy
import nltk
from nltk.corpus import stopwords
!python -m spacy download en_core_web_sm
nlp = spacy.load("en_core_web_sm")
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
real_news_df = pd.read_csv("/content/sample_data/gossipcop_real.csv")
fake_news_df = pd.read_csv("/content/sample_data/gossipcop_fake.csv")
def preprocess_text(text):
    if not isinstance(text, str):
        return ''
    text = text.lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    doc = nlp(text)
    words = [token.lemma_ for token in doc if token.is_alpha and token.text not in stop_words]
    return ' '.join(words)
real_news_df['label'] = 1
fake_news_df['label'] = 0
combined_df = pd.concat([real_news_df, fake_news_df], ignore_index=True)
combined_df['cleaned_title'] = combined_df['title'].apply(preprocess_text)
print(combined_df[['id', 'news_url', 'title', 'cleaned_title', 'label']].head())
```

Expanding Contractions:

```python
import re
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
contractions_dict = {
    "can't": "cannot",
    "won't": "will not",
    "n't": " not",
    "'re": " are",
    "'s": " is",
    "'d": " would",
    "'ll": " will",
    "'t": " not",
    "'ve": " have",
    "'m": " am"
}
def expand_contractions(text, contractions_dict):
    if not isinstance(text, str):
        return ''
    contractions_re = re.compile('(%s)' % '|'.join(re.escape(key) for key in contractions_dict.keys()))
    def replace(match):
        return contractions_dict[match.group(0)]
    return contractions_re.sub(replace, text)
real_news_df = pd.read_csv("/content/sample_data/gossipcop_real.csv")
fake_news_df = pd.read_csv("/content/sample_data/gossipcop_fake.csv")
real_news_df['label'] = 1
fake_news_df['label'] = 0
combined_df['expanded_title'] = combined_df['title'].apply(lambda x: expand_contractions(x, contractions_dict))
G = nx.DiGraph()
nodes = [
    "Input Data (combined_df['title'])",
    "Contractions Dictionary (contractions_dict)",
    "Expand Contractions Function (expand_contractions)",
    "Output Data (combined_df['expanded_title'])"
]
G.add_nodes_from(nodes)
edges = [
    ("Input Data (combined_df['title'])", "Expand Contractions Function (expand_contractions)"),
    ("Contractions Dictionary (contractions_dict)", "Expand Contractions Function (expand_contractions)"),
    ("Expand Contractions Function (expand_contractions)", "Output Data (combined_df['expanded_title'])")
]
G.add_edges_from(edges)
pos = nx.spring_layout(G)
plt.figure(figsize=(12, 7))
nx.draw(
    G, pos, with_labels=True, node_size=4000, node_color='lightblue',
    font_size=10, font_weight='bold', edge_color='gray', arrowsize=20
)
plt.title("Data Flow Graph for Contraction Expansion Process")
plt.show()
print(combined_df[['id', 'title', 'expanded_title', 'label']].head())
```

Downsampling:

```python
import pandas as pd
from sklearn.utils import resample
real_df = pd.read_csv("/content/sample_data/gossipcop_real.csv")
fake_df = pd.read_csv("/content/sample_data/gossipcop_fake.csv")
real_df['label'] = 1
fake_df['label'] = 0
combined_df = pd.concat([real_df, fake_df], ignore_index=True)
df_majority = combined_df[combined_df['label'] == 1]
df_minority = combined_df[combined_df['label'] == 0]
df_majority_downsampled = resample(df_majority,
                                   replace=False,
                                   n_samples=len(df_minority),
                                   random_state=42)
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
df_downsampled = df_downsampled.sample(frac=1, random_state=42).reset_index(drop=True)
print("Class distribution after downsampling:")
print(df_downsampled['label'].value_counts())
df_downsampled.to_csv('downsampled_combined.csv', index=False)
```

Data Augmentation

Synonym Replacement:

```python
import pandas as pd
import random
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')
real_df = pd.read_csv("/content/sample_data/gossipcop_real.csv")
fake_df = pd.read_csv("/content/sample_data/gossipcop_fake.csv")
combined_df = pd.concat([real_df, fake_df], ignore_index=True)
def synonym_replacement(text, num_replacements=1):
    words = word_tokenize(text)
    new_words = words.copy()
    random.shuffle(words)
    replacements = 0
    for word in words:
        synonyms = wordnet.synsets(word)
        synonyms = [lemma.name() for syn in synonyms for lemma in syn.lemmas() if lemma.name().lower() != word.lower()]
        if synonyms:
            synonym = random.choice(synonyms)
            new_words = [synonym if w == word else w for w in new_words]
            replacements += 1
        if replacements >= num_replacements:
            break
    return " ".join(new_words)
combined_df['synonym_replaced_title'] = combined_df['title'].dropna().apply(lambda x: synonym_replacement(x))
print(combined_df[['title', 'synonym_replaced_title']].head())
combined_df.to_csv('combined_synonym_replaced.csv', index=False)
```

Random Insertion/Deletion/Swap:

```python
import pandas as pd
import random
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
real_df = pd.read_csv("/content/sample_data/gossipcop_real.csv")
fake_df = pd.read_csv("/content/sample_data/gossipcop_fake.csv")
combined_df = pd.concat([real_df, fake_df], ignore_index=True)
def random_insertion(text, num_insertions=1):
    words = word_tokenize(text)
    for _ in range(num_insertions):
        random_word = random.choice(words)
        insert_index = random.randint(0, len(words))
        words.insert(insert_index, random_word)
    return ' '.join(words)
def random_deletion(text, num_deletions=1):
    words = word_tokenize(text)
    for _ in range(num_deletions):
        if len(words) == 0:
            break
        delete_index = random.randint(0, len(words)-1)
        words.pop(delete_index)
    return ' '.join(words)
def random_swap(text, num_swaps=1):
    words = word_tokenize(text)
    for _ in range(num_swaps):
        if len(words) < 2:
            break
        swap_idx1, swap_idx2 = random.sample(range(len(words)), 2)
        words[swap_idx1], words[swap_idx2] = words[swap_idx2], words[swap_idx1]
    return ' '.join(words)
combined_df['random_insertion_title'] = combined_df['title'].dropna().apply(lambda x: random_insertion(x, num_insertions=1))
combined_df['random_deletion_title'] = combined_df['title'].dropna().apply(lambda x: random_deletion(x, num_deletions=1))
combined_df['random_swap_title'] = combined_df['title'].dropna().apply(lambda x: random_swap(x, num_swaps=1))
print(combined_df[['title', 'random_insertion_title', 'random_deletion_title', 'random_swap_title']].head())
combined_df.to_csv('combined_random_augmentation.csv', index=False)
```

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
from collections import Counter
real = pd.read_csv("/content/sample_data/gossipcop_real.csv")
fake = pd.read_csv("/content/sample_data/gossipcop_fake.csv")
real['label'] = 'Real'
fake['label'] = 'Fake'
combined = pd.concat([real, fake]).dropna(subset=['title'])
def handle_numerics(text):
    if not isinstance(text, str):
        return text
    processed_text = re.sub(r'\b\d+\b', 'NUM', text)
    return processed_text
combined['handled_numerics_title'] = combined['title'].apply(lambda x: handle_numerics(x))
def get_word_counts(text_series):
    words = ' '.join(text_series.dropna()).split()
    return Counter(words)
original_word_counts = get_word_counts(combined['title'])
processed_word_counts = get_word_counts(combined['handled_numerics_title'])
original_words, original_counts = zip(*original_word_counts.most_common(10))
processed_words, processed_counts = zip(*processed_word_counts.most_common(10))
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.barh(original_words, original_counts, color='skyblue')
plt.title('Top 10 Words in Original Titles')
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.subplot(1, 2, 2)
plt.barh(processed_words, processed_counts, color='orange')
plt.title('Top 10 Words in Processed Titles (Numerics Handled)')
plt.xlabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import spacy
import matplotlib.pyplot as plt
from collections import Counter
nlp = spacy.load("en_core_web_sm")
real = pd.read_csv("/content/sample_data/gossipcop_real.csv")
fake = pd.read_csv("/content/sample_data/gossipcop_fake.csv")
real['label'] = 'Real'
fake['label'] = 'Fake'
combined = pd.concat([real, fake]).dropna(subset=['title'])
def get_pos_tags_batch(titles):
    docs = list(nlp.pipe(titles, disable=["ner", "parser"]))
    return [[token.pos_ for token in doc] for doc in docs]
titles = combined['title'].tolist()
pos_tags_batch = get_pos_tags_batch(titles)
combined['pos_tags'] = pos_tags_batch
all_pos_tags = [tag for tags_list in combined['pos_tags'] for tag in tags_list]
pos_counts = Counter(all_pos_tags)
plt.figure(figsize=(10, 6))
plt.bar(pos_counts.keys(), pos_counts.values(), color='teal')
plt.title('Distribution of Parts of Speech (POS) Tags in Titles')
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

Named Entity Recognition (NER):

```python
import spacy
import pandas as pd

# Load the pre-trained spaCy model (English)
nlp = spacy.load("en_core_web_sm")

# Function to perform NER on text data
def extract_named_entities(text):
    if not isinstance(text, str):
        return []
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

# Apply NER to the 'title' column of both datasets
real_news_df['named_entities'] = real_news_df['title'].apply(extract_named_entities)
fake_news_df['named_entities'] = fake_news_df['title'].apply(extract_named_entities)

# Display a sample of the resulting dataframes
print("Real News DataFrame with Named Entities:")
print(real_news_df[['title', 'named_entities']].head())

print("\nFake News DataFrame with Named Entities:")
print(fake_news_df[['title', 'named_entities']].head())
```
