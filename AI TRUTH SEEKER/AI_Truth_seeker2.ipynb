```python
# Initialize the stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Define a preprocessing function
def preprocess_text(title):
    # Tokenization
    tokens = word_tokenize(title.lower())  # Convert to lowercase and tokenize
    # Stemming
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    # Lemmatization
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in stemmed_tokens]
    return " ".join(lemmatized_tokens)
```

## NER Named Entity Recognition

```python
from nltk import word_tokenize
from nltk import pos_tag, ne_chunk
from nltk.tree import Tree
```

```python
# Function to perform Named Entity Recognition (NER)
def perform_ner(text):
    # Tokenize the text
    tokens = word_tokenize(text)
    # Perform POS tagging
    pos_tags = pos_tag(tokens)
    # Perform Named Entity Recognition
    named_entities = ne_chunk(pos_tags)

    # Extract named entities from the chunked structure
    entities = []
    for chunk in named_entities:
        if isinstance(chunk, Tree):  # If the chunk is a named entity
            entity_name = " ".join(c[0] for c in chunk)
            entity_type = chunk.label()  # Get the type of entity (e.g., PERSON, ORGANIZATION)
            entities.append((entity_name, entity_type))

    return entities
```

```python
def visualize_entities(text):
    # Tokenize and tag the text
    tokens = word_tokenize(df1['text'][0])
    tagged_tokens = pos_tag(tokens)

    # Use ne_chunk to get named entities
    named_entities = ne_chunk(tagged_tokens)

    # Define colors for entity types
    entity_colors = {
        "PERSON": "green",
        "ORGANIZATION": "cyan",
        "GPE": "yellow",
        "LOCATION": "magenta",
        "DATE": "blue",
        "MONEY": "red",
    }

    # Parse and display entities with color
    for chunk in named_entities:
        if hasattr(chunk, 'label'):
            entity_type = chunk.label()
            entity_name = " ".join(c[0] for c in chunk)
            color = entity_colors.get(entity_type, "white")
            print(colored(f"{entity_name} ({entity_type})", color))
        else:
            # Print non-entity words normally
            print(chunk[0], end=" ")
    print()  # Add a newline at the end

# Test the function with a sample sentence

visualize_entities(df1['text'][0])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from gensim.models import Word2Vec
import numpy as np
from tensorflow.keras.utils import to_categorical
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(train_texts)
```

```python
train_sequences=tokenizer.texts_to_sequences(train_texts)
test_sequences=tokenizer.texts_to_sequences(test_texts)
```

```python
word_index=tokenizer.word_index
```

```python
word_index
```

```python
max_sequence_length=max(len(seq) for seq in train_sequences)
```

```python
max_sequence_length
```

```python
train_padded=pad_sequences(train_sequences,maxlen=max_sequence_length,padding="post")
test_padded=pad_sequences(test_sequences,maxlen=max_sequence_length,padding="post")
```

```python
train_padded
```

```python
train_labels=to_categorical(train_labels)
```

```python
train_labels
```

```python
test_labels=to_categorical(test_labels)
```

```python
w2v_model=Word2Vec(sentences=[text.split() for text in train_texts], vector_size=100)
```

```python
embedding_dim=100
embedding_matrix=np.zeros((len(word_index)+1,embedding_dim))
```

```python
embedding_matrix
```

```python
for word, i in word_index.items():
  if word in w2v_model.wv:
    embedding_matrix[i]=w2v_model.wv[word]
```

```python
embedding_matrix
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM, Dense
```

```python
model=Sequential([
    Embedding(input_dim=len(word_index)+1,
              output_dim=embedding_dim,
              weights=[embedding_matrix],
              input_length=max_sequence_length,
              trainable=False
              ),
    LSTM(128),
    Dense(2,activation="softmax")
])
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```
